```python
import pandas as pd
import os
```

```python
file_prefix = "results"

directory = 'C:\EPO-Project\Results'

files = [file for file in os.listdir(directory) if file.startswith(file_prefix)]
```

```python
files
```

```python
result_df = pd.DataFrame()

for file in files:
    
    file_path = os.path.join(directory, file)

    file_df = pd.read_pickle(file_path)

    file_split = file.split('_')

    experiment = file_split[3]
    opposition = file_split[5]
    algorithm = file_split[7].split('\'')[1]
    input_rep = file_split[9]

    dictionary = {'Experiment':[experiment],'Opposition':[opposition],'Algorithm':[algorithm],'Input':[input_rep],'Data':[file_df]}

    result_df = pd.concat([result_df,pd.DataFrame(dictionary)],ignore_index=False)

print(result_df)
```

```python
def extract_info(df):

    result_df = pd.DataFrame()

    for i in df.iterrows():

        test_acc = i[1]['Data']['test_Accuracy'][0].mean()
        test_acc_std = i[1]['Data']['test_Accuracy'][0].std()
        test_f1 = i[1]['Data']['test_F1'][0].mean()
        test_f1_std = i[1]['Data']['test_F1'][0].std()
        test_mcc = i[1]['Data']['test_MCC'][0]
        test_mcc_std = i[1]['Data']['test_MCC'][0].std()

        algo = i[1]['Algorithm']
        input_rep = i[1]['Input']
        
        #FINISH DATAFRAME
        dictionary = {'Algorithm':[algo],'Input':[input_rep],'Acc':[test_acc],'Acc_std':[test_acc_std]}
        result_df = pd.concat([result_df,pd.DataFrame(dictionary)],ignore_index=False)
    
    #ORDER BY CATEGORIES

    # Define the order of categories
    category_order = ['A', 'B', 'C']

    # Convert the 'Category' column to a Categorical type with specified order
    df['Category'] = pd.Categorical(df['Category'], categories=category_order, ordered=True)

    # Sort the DataFrame by the categorical column
    df_sorted_categorical = df.sort_values(by='Category')

    return result_df
```
